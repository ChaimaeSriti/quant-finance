# Time Series Analysis
## Smoothing methods

```python
import pandas as pd

from models import SmoothingMethods
```

```python
# import data
data = pd.read_csv('data/^TNX.csv', index_col=['Date'], usecols=['Date', 'Open'], skiprows=range(1, 2000),
                   parse_dates=['Date'])
```

## Moving Average

```python
# Initialize Moving average
moving_avg = SmoothingMethods.MovingAverage(data)

# Set number of observations
moving_avg.simple(window=20)

# plot results
moving_avg.plot_simple(intervals=True, anomalies=True)
```

## Exponential Smoothing

```python
# Initialize Exponential Moving average
exp_avg = SmoothingMethods.ExponentialMovingAverage(data)

# Set smoothing factor
exp_avg.basic_smoothing(alpha=0.05)

# Plot results
exp_avg.plot_basic()
```

## Double Exponential Smoothing

```python
# Set smoothing factor (alpha) and trend (beta)
exp_avg.double_smoothing(alpha=0.05, beta=0.5)

# Plot results
exp_avg.plot_double()
```
